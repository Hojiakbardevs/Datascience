```python
import numpy as np
import matplotlib.pyplot as plt

# Trapetsiya shaklidagi a'zolik funksiyasi
def trapezoidal_membership(x, a, b, c, d):
    return np.maximum(0, np.minimum(1, np.minimum((x-a)/(b-a), (d-x)/(d-c))))

# Bo'y, vazn va yosh a'zolik funksiyalari uchun trapetsiya parametrlarini belgilash
a_boy, b_boy, c_boy, d_boy = 160, 170, 180, 190
a_vazn, b_vazn, c_vazn, d_vazn = 60, 70, 80, 90
a_yosh, b_yosh, c_yosh, d_yosh = 25, 30, 40, 45

# X qiymatlari
x_values = np.linspace(0, 200, 1000)

# A'zolik qiymatlari
membership_boy = trapezoidal_membership(x_values, a_boy, b_boy, c_boy, d_boy)
membership_vazn = trapezoidal_membership(x_values, a_vazn, b_vazn, c_vazn, d_vazn)
membership_yosh = trapezoidal_membership(x_values, a_yosh, b_yosh, c_yosh, d_yosh)

# Kesishmani aniqlash
intersection = np.minimum(np.minimum(membership_boy, membership_vazn), membership_yosh)

# Grafikni chizish
plt.figure(figsize=(12, 6))
plt.plot(x_values, membership_boy, label="Bo'y")
plt.plot(x_values, membership_vazn, label="Vazn")
plt.plot(x_values, membership_yosh, label="Yosh")
plt.plot(x_values, intersection, 'r--', label="Kesishma")

# Noravshan to'plamlarni ko'rsatish
plt.fill_between(x_values, membership_boy, alpha=0.2, label="Bo'y noravshan to'plami")
plt.fill_between(x_values, membership_vazn, alpha=0.2, label="Vazn noravshan to'plami")
plt.fill_between(x_values, membership_yosh, alpha=0.2, label="Yosh noravshan to'plami")
plt.fill_between(x_values, intersection, color='red', alpha=0.3, label="Kesishma")

# Grafik sozlamalari
plt.title("Bo'y, vazn va yosh noravshan to'plamlari va ularning kesishmasi")
plt.xlabel("Qiymatlar")
plt.ylabel("A'zolik darajasi")
plt.grid(True)
plt.legend()

# Grafikni ko'rsatish
plt.show()

# Kesishma to'plamining xususiyatlarini aniqlash
max_intersection = np.max(intersection)
x_max_intersection = x_values[np.argmax(intersection)]

print(f"Kesishmaning maksimal qiymati: {max_intersection:.4f}")
print(f"Maksimal kesishma nuqtasi: {x_max_intersection:.2f}")

# Alpha-kesim
alpha = 0.1
alpha_cut = x_values[(intersection >= alpha)]
if len(alpha_cut) > 0:
    print(f"{alpha}-kesim intervali: [{alpha_cut[0]:.2f}, {alpha_cut[-1]:.2f}]")
else:
    print(f"{alpha}-kesim mavjud emas")

# Noravshan to'plamlarning o'zaro kesishuv darajasini hisoblash
intersection_area = np.trapz(intersection, x_values)
union_area = np.trapz(np.maximum(np.maximum(membership_boy, membership_vazn), membership_yosh), x_values)
similarity = intersection_area / union_area

print(f"Noravshan to'plamlarning o'xshashlik darajasi: {similarity:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Tasodifiy bo'y ma'lumotlarini yaratish
np.random.seed(42)  # Takrorlanuvchi natijalar uchun
num_people = 1000
heights = np.random.uniform(150, 200, num_people)

# Bo'y uchun noravshan to'plamlar
def short_membership(x):
    return np.maximum(0, np.minimum(1, (170 - x) / 20))

def medium_membership(x):
    return np.maximum(0, np.minimum(np.minimum((x - 160) / 15, (185 - x) / 15), 1))

def tall_membership(x):
    return np.maximum(0, np.minimum(1, (x - 175) / 20))

# Noravshan to'plamlarni hisoblash
short = short_membership(heights)
medium = medium_membership(heights)
tall = tall_membership(heights)

# Gistogramma va noravshan to'plamlarni chizish
plt.figure(figsize=(12, 6))

# Gistogramma
plt.hist(heights, bins=30, density=True, alpha=0.6, color='gray', label='Bo\'y taqsimoti')

# Noravshan to'plamlar
x = np.linspace(150, 200, 200)
plt.plot(x, short_membership(x), 'r-', label='Past bo\'yli')
plt.plot(x, medium_membership(x), 'g-', label='O\'rta bo\'yli')
plt.plot(x, tall_membership(x), 'b-', label='Baland bo\'yli')

plt.title("Bo'y taqsimoti va noravshan to'plamlar")
plt.xlabel("Bo'y (sm)")
plt.ylabel("Zichlik / A'zolik darajasi")
plt.legend()
plt.grid(True)
plt.show()

# Statistik ma'lumotlar
print(f"O'rtacha bo'y: {np.mean(heights):.2f} sm")
print(f"Median bo'y: {np.median(heights):.2f} sm")
print(f"Standart og'ish: {np.std(heights):.2f} sm")

# Noravshan to'plamlarga a'zolik
print("\nNoravshan to'plamlarga a'zolik:")
print(f"Past bo'yli: {np.mean(short):.2%}")
print(f"O'rta bo'yli: {np.mean(medium):.2%}")
print(f"Baland bo'yli: {np.mean(tall):.2%}")

# Alpha-kesim misollar
alpha = 0.5
print(f"\n{alpha}-kesim intervallari:")
print(f"Past bo'yli: {x[short_membership(x) >= alpha].min():.2f} - {x[short_membership(x) >= alpha].max():.2f} sm")
print(f"O'rta bo'yli: {x[medium_membership(x) >= alpha].min():.2f} - {x[medium_membership(x) >= alpha].max():.2f} sm")
print(f"Baland bo'yli: {x[tall_membership(x) >= alpha].min():.2f} - {x[tall_membership(x) >= alpha].max():.2f} sm")

# Noravshan to'plamlar kesishmasi
intersection = np.minimum(np.minimum(short_membership(x), medium_membership(x)), tall_membership(x))
plt.figure(figsize=(12, 6))
plt.plot(x, short_membership(x), 'r-', label='Past bo\'yli')
plt.plot(x, medium_membership(x), 'g-', label='O\'rta bo\'yli')
plt.plot(x, tall_membership(x), 'b-', label='Baland bo\'yli')
plt.plot(x, intersection, 'k--', label='Kesishma')
plt.fill_between(x, intersection, alpha=0.3)
plt.title("Noravshan to'plamlar va ularning kesishmasi")
plt.xlabel("Bo'y (sm)")
plt.ylabel("A'zolik darajasi")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Bo'y uchun noravshan to'plamlar
def short_membership(x):
    return np.maximum(0, np.minimum(1, (170 - x) / 20))

def medium_membership(x):
    return np.maximum(0, np.minimum(np.minimum((x - 160) / 15, (185 - x) / 15), 1))

def tall_membership(x):
    return np.maximum(0, np.minimum(1, (x - 175) / 20))

# Bo'y qiymatlari
x = np.linspace(150, 200, 200)

# Noravshan to'plamlarni hisoblash
short = short_membership(x)
medium = medium_membership(x)
tall = tall_membership(x)

# Noravshan to'plamlar kesishmasi
intersection_short_medium = np.minimum(short, medium)
intersection_medium_tall = np.minimum(medium, tall)
intersection_short_tall = np.minimum(short, tall)
intersection_all = np.minimum(np.minimum(short, medium), tall)

# Grafikni chizish
plt.figure(figsize=(12, 8))

plt.plot(x, short, 'r-', label='Past bo\'yli')
plt.plot(x, medium, 'g-', label='O\'rta bo\'yli')
plt.plot(x, tall, 'b-', label='Baland bo\'yli')

plt.plot(x, intersection_short_medium, 'm--', label='Past-O\'rta kesishmasi')
plt.plot(x, intersection_medium_tall, 'c--', label='O\'rta-Baland kesishmasi')
plt.plot(x, intersection_short_tall, 'y--', label='Past-Baland kesishmasi')
plt.plot(x, intersection_all, 'k--', label='Uchala to\'plam kesishmasi')

plt.fill_between(x, intersection_short_medium, alpha=0.1, color='magenta')
plt.fill_between(x, intersection_medium_tall, alpha=0.1, color='cyan')
plt.fill_between(x, intersection_short_tall, alpha=0.1, color='yellow')
plt.fill_between(x, intersection_all, alpha=0.2, color='black')

plt.title("Noravshan to'plamlar va ularning kesishmalari")
plt.xlabel("Bo'y (sm)")
plt.ylabel("A'zolik darajasi")
plt.legend()
plt.grid(True)
plt.show()

# Kesishmalar haqida ma'lumot
print("Kesishmalar haqida ma'lumot:")
print(f"Past-O'rta max kesishmasi: {np.max(intersection_short_medium):.2f}")
print(f"O'rta-Baland max kesishmasi: {np.max(intersection_medium_tall):.2f}")
print(f"Past-Baland max kesishmasi: {np.max(intersection_short_tall):.2f}")
print(f"Uchala to'plam max kesishmasi: {np.max(intersection_all):.2f}")
```
