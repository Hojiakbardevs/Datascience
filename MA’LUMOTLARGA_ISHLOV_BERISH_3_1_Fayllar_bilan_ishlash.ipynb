```python
import pandas as pd
import numpy as np
import sqlite3 as sql
```

```python
from urllib.request import urlretrieve
url = "https://github.com/anvarnarz/praktikum_datasets/raw/main/electricity.db"
urlretrieve(url,'electricity.db') # faylni yuklab olish
```

```python
db = sql.connect("electricity.db")
cursor = db.cursor()
```

```python
cursor.execute("SELECT name FROM  sqlite_master WHERE type = 'table'")
print(cursor.fetchall())
```

```python
cursor.execute("PRAGMA table_info('elect')")
cursor.fetchall()
```

```python
pd.read_sql("SELECT * FROM elect LIMIT 15", db, index_col='index')
```

```python
query = "SELECT DISTINCT country FROM elect"
countries = pd.read_sql(query,db)
countries
```

```python
"Uzbekistan" in countries['country'].values
```

```python
query = "SELECT * FROM elect WHERE country='Uzbekistan'"
pd.read_sql(query,db)
```

```python
query = "SELECT * FROM elect WHERE country='Uzbekistan' AND year>2010"
pd.read_sql(query,db)
```

```python
query = "SELECT * FROM elect WHERE country='Uzbekistan' AND year BETWEEN 2000 AND 2010 "
pd.read_sql(query,db)
```

```python
query = "SELECT * FROM elect WHERE country='Uzbekistan' AND year IN (2004, 2003) "
pd.read_sql(query,db)
```

```python
query = "SELECT * FROM elect WHERE country LIKE '%stan' AND year BETWEEN 2000 AND 2010 "
pd.read_sql(query,db)
```
